```python
from io import BytesIO
import gzip
import ssl
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
from collections import OrderedDict
ssl._create_default_https_context = ssl._create_unverified_context
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://nubs.nju.edu.cn/8878/list.htm"
```

```python
html = urlopen(url, )
html
```

```python
html_text = bytes.decode(html.read())
html_text
```

```python
soup = BeautifulSoup(html_text, 'html.parser')
soup
```

```python
soup.find()
```

```python
blocks = soup.find('div','news_box')
blocks
```

```python
soup.find_all("div", "news_list main")
```

```python
blocks = soup.find('div','news_list main').find_all("li")
```

```python
result = []
for block in blocks:
    mydict = OrderedDict()
    mydict['name']= block.find("li").text
    
    result.append(mydict)
df = pd.DataFrame(result)
df
```

```python
df.index = range(1,len(df.index)+1)
df
```

```python
import ssl
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
from collections import OrderedDict
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.sustech.edu.cn/zh/colleges/mathematics.html"
```

```python
html = urlopen(url)
```

```python
html_text = bytes.decode(html.read())
html_text
```

```python
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
blocks = soup.find('div','w list-main clearfix').find_all("li")
```

```python
result = []
for block in blocks:
    mydict = OrderedDict()
    mydict['name']= block.find("div", 'name').text
    mydict['pos']= block.find("div", 'p').text
    mydict['dep']= block.find("div", 'dep').text
    result.append(mydict)
df = pd.DataFrame(result)
df
```

```python
df.index = range(1,len(df.index)+1)
df
```
